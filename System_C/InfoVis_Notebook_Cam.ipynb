```python
import pandas as pd
import numpy as np
import altair as alt

df = pd.read_csv(r'C:\Users\cswat\OneDrive\Documents\InformationVisData\WW1_BOMBING_DATA-CLEANED.csv', parse_dates=['MSNDATE'], dayfirst=True)

df_count = df.groupby([df['MSNDATE'].dt.date, 'MDS']).size().reset_index(name='Counts')
df_count['MSNDATE'] = pd.to_datetime(df_count['MSNDATE'])

selection = alt.selection_point(fields=['MDS'], on='click', bind='legend')

chart_bar = alt.Chart(df_count).mark_bar().encode(
    x=alt.X('MSNDATE:T', title='Date'),
    y=alt.Y('Counts:Q', title='Number of Missions'),
    color=alt.condition(selection, 'MDS:N', alt.value('lightgray'), legend=alt.Legend(title="Model of Plane")),
    tooltip=['MDS:N', 'Counts:Q', 'MSNDATE:T']
).add_params(selection)

chart_bar
```
